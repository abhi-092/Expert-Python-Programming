```python
def decor(func):
  def wrapper():
    func()

    print("Start")
  return wrapper


@decor
def printer():
  print("Hello")
  print("2Hello")

printer()
```

```python
def decor(func):
  def inner():
    func()
    print("1welcome")
  return inner

@decor
def printer():
  print("2welcome")
  print("3welcome")


printer()
```

```python
def decor(func):
  def wrapper():
    print('------')
    func()
    print('------')
  return wrapper

@decor
def printer():
  print('Hello')

printer()
```
