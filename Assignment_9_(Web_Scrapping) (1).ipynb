```python
url = "https://www.scrapethissite.com/pages/simple/"
response=requests.get(url)

if response.status_code == 200:
  print("Success")

else:
  print("Failed")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
country_name = soup.find_all("h3", class_ = "country-name")
capitals = soup.find_all("span", class_ = "country-capital")

for name, capital in zip(country_name, capitals):
  print(f"{name.text} - {capital.text}")
```

```python
url = "https://www.goodreads.com/"

header = {"user-Agent": "Mozilla/5.0"}
response = requests.get(url, headers = header)

if response.status_code == 200:
  print("Success")

else:
  print("Failed")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
quotes = soup.find_all("div", class_ = "quoteText")
authors = soup.find_all("span", class_ = "authorOrTitle")

for quote, author in zip(quotes, authors):
  print(f"{author.text} - {quote.text}")
```

```python
  import requests
  from bs4 import BeautifulSoup

  url = "https://www.scrapethissite.com/pages/simple/"
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  rk = soup.title
  print(rk.text)
  countries = soup.find_all("div", class_="country")
  all_data = []

  print(f"Scraping {len(countries)} countries...")

  for i, country in enumerate(countries, 1):
      name = country.find("h3", class_="country-name").get_text(strip=True)
      capital = country.find("span", class_="country-capital").get_text(strip=True)
      population = country.find("span", class_="country-population").get_text(strip=True)
      area = country.find("span", class_="country-area").get_text(strip=True)

      data = {"name": name, "capital": capital, "population": population, "area": area}
      all_data.append(data)

      print(f"{i}. {name} | Capital: {capital} | Pop: {population} | Area: {area}")


  print(f"\nTotal scraped: {len(all_data)} countries")
```
