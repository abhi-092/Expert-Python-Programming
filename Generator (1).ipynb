```python
def my_gen():
    yield "Abhi"
    yield 2
    yield 3

gen = my_gen()
print(next(gen))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    yield 4

gen = my_generator()

for num in gen:
    print(num)
```

```python
def fibonacci(n):
    a, b = 0, 1
    for n in range(n):
        yield a
        a, b = b, a + b

fib = fibonacci(13)
print(list(fib))
```

```python
#Question: Generator to yield numbers from 1 to n
def gen_numbers(n):
    for i in range(1, n+1):
        yield i
```

```python
#Question: Generator for odd numbers up to n
def odd_numbers(n):
    for i in range(1, n+1, 2):
        yield i
```

```python
#Question: Yield characters of string in reverse.
def reverse_str(s):
    for char in s[::-1]:
        yield char
```

```python
#Question: Infinite generator of even numbers.
def infinite_even():
    n = 0
    while True:
        yield n
        n += 2
```

```python
#Question: Cubes
def cubes(n):
    for i in range(1, n+1):
        yield i**3
```
