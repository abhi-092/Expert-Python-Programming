***Ques_04. Implement a Musician class and create another child class of your choice that extends an Artist class. Include an additional attribute, instrument_played.***

```python
class Musician:
  def __init__(self,name,age):
    self.name = name
    self.age = age

class Artist(Musician):
  def __init__(self,name,age,instrument_played):
    super().__init__(name,age)
    self.instrument_played = instrument_played

music1 = Artist('Ram','20','Tabla')
print('the artist name is',music1.name,'his age',music1.age,'and played',music1.instrument_played)
```

***Ques_05. Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute, genre.***

```python
class FictionBook:
  def __init__(self,writer,name):
    self.writer = writer
    self.name = name

class Book(FictionBook):
  def __init__(self,writer,name,genre):
    super().__init__(writer,name)
    self.genre = genre

fiction1 = Book('Vikas','Technical','Guide')
print(fiction1.writer,fiction1.name,fiction1.genre)
```

***BonusQues:- Create a student class that takes the name & marks of 3 subjects as arguments in the constructor. Then create a method to print the average.***

```python
class Student:
  def __init__(self,name,marks1,marks2,marks3):
    self.name = name
    self.marks1 = marks1
    self.marks2 = marks2
    self.marks3 = marks3

  def get_average(self):
    total_sum = self.marks1 + self.marks2 + self.marks3
    average = total_sum / 3
    return average

student1 = Student('Vikas',78,66,90)
print(student1.name,student1.get_average())
```
